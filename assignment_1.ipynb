Ques1. What is data cleaning, and why is it important in data analysis?
 -Data cleaning (also called data cleansing or data preprocessing) is the process of identifying and correcting errors, inconsistencies, inaccuracies, or missing values in a dataset before performing analysis.

A. What are the potential consequences of analyzing unclean or messy data?
-1. Incorrect Conclusions
-2. poor business decisions
-3. Reduced Model Accuracy
-4. Misleading visualizations
-5. Wasted time and resources
-6. Loss of Credibility
-7. Compliance and Legal Risks
B. Explain the common steps involved in cleaning and organizing data.
- Data Inspection
- Remove Duplicate Records
- Handle Missing Data
- Correct Errors and Inconsistencies
- Standardize Formats
- Validate Data Types
- Detect and Handle Outliers
- Organize and Structure the Data
- Document Changes

ques3. Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning?
- TRIM Removes extra spaces from text, leaving only single spaces between words.
-LEFT Extracts a specified number of characters from the start of a text string.
- RIGHT Extracts characters from the end of a text string.
- MID Extracts text from a specific position inside a string.
-CONCAT Joins text from multiple cells into one.
